```python
import requests # request http, api
import pandas as pd # tabluar data
import plotly.express as px 
from census import Census
from us import states
```

```python
# data from census api 
```

```python
#get api request from census 

url = "https://api.census.gov/data/2019/pep/charage?get=NAME,POP&for=us:*&key=#####"
response = requests.request("GET", url)
```

```python
response  # check to make sure the call worked
```

```python
response.text
```

```python
# function to convert json to dataframe

def json_to_dataframe(response):
    """
    Convert response to dataframe
    """
    return pd.DataFrame(response.json()[1:], columns=response.json()[0])
```

```python
json_to_dataframe(response)
```

```python
# population united states by state

url2 = "https://api.census.gov/data/2019/pep/charage?get=NAME,POP&for=state:*&key=#####"
response = requests.request("GET", url2)
```

```python
response
```

```python
#dataframe with population of each state

state_pop = json_to_dataframe(response)
state_pop.head(25)
```

```python
# convert population to int

state_pop['POP'] = state_pop['POP'].astype(int)
state_pop.info()
```
